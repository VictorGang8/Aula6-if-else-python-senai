Atividade 1 — Classificação de número

Objetivo: usar if e else

Crie um programa que:

Peça um número

Verifique se ele é:

Positivo

Negativo



```python
#atividade 1
numero = float(input("Digite um número: "))
if numero > 0:
    print("O número é positivo.")
elif numero == 0:
    print("O número é nulo.")
else:
    print("O número é negativo.")
```

Atividade 2 — Par ou Ímpar

Objetivo: introduzir lógica com operador

Crie um programa que:

Peça um número

Verifique se ele é:

Par

Ímpar

Dica: usar % 2

```python
#atividade 2
numero = int(input("Digite um número: "))
if numero % 2 == 0:
    print("O número é par.")
elif numero == 0:
    print("O número é nulo.")
else:
    print("O número é ímpar.")
```

Atividade 3 — Sistema de notas

Objetivo: usar elif

Crie um programa que:

Peça a nota de um aluno

Classifique:
Nota

Resultado >= 7 Aprovado



Recuperação >= 5



Reprovado  < 5

```python
#Atividade 03
nota = float(input("Digite a nota do aluno: "))
if nota >= 7:
  print("Aluno Aprovado com a nota: {}".format(nota))
elif nota >= 5:
  print("Aluno de Recuperação com a nota: {}".format(nota))
else:
  print("Aluno Reprovado com a nota: {}".format(nota))
```

Atividade 4 — Simulação de login simples

Objetivo: aplicar condição mais real

Crie um programa que:

Peça:

usuário

senha

Verifique:

usuario correto: admin
senha correta: 1234

Resultado:

Se ambos corretos → "Login realizado"

Senão → "Usuário ou senha inválidos“

```python
usuario = input("Login:")
senha = input("Senha:")
if usuario == "admin" and senha == "1234":
    print('Login realizado')
elif usuario != "admin":
    print("Usuário incorreto")
elif senha != "1234":
    print("Senha incorreta")
else:
  print('Usuário ou senha inválidos!')

```
